```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Data
combinations = {
    'S1-R1': 2.3,
    'S2-R1': 4.0,
    'S1-R2': 3.7,
    'S2-R2': 3.3,
    'S1-R3': 7.3,
    'S2-R3': 7.8,
    'S1-R4': 6.8,
    'S2-R4': 8.0,
    'S1-R5': 7.2,
    'S2-R5': 9.0,
    'S1-R6': 4.7,
    'S2-R6': 6.8
}

# Convert data to a matrix format
matrix = pd.DataFrame(index=['S1', 'S2'], columns=['R1', 'R2', 'R3', 'R4', 'R5', 'R6'])

for combo, dist in combinations.items():
    s, r = combo.split('-')
    matrix.at[s, r] = dist

# Convert to float and fill NaN values
matrix = matrix.astype(float).fillna(0)

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(matrix, annot=True, cmap="YlGnBu", cbar_kws={'label': 'Distance (m)'})
plt.title('Distance between Sources and Receivers')
plt.show()
```
